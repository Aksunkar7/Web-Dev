<h1 align="center">Define a Route</h1>

```python
// app.routes.ts
import {Routes} from '@angular/router';

import {Home} from './home/home';
import {User} from './user/user';

export const routes: Routes = [
  {
    path: '',
    title: 'App Home Page',
    component: Home,
  },
  {
    path: 'user',
    title: 'App User Page',
    component: User,
  },
];
```

```python
// home/home.ts
import {Component} from '@angular/core';

@Component({
  selector: 'app-home',
  template: ` <div>Home Page</div> `,
})
export class Home {}
```

```python
// user/user.ts
import {Component} from '@angular/core';

@Component({
  selector: 'app-user',
  template: ` <div>Username: {{ username }}</div> `,
})
export class User {
  username = 'youngTech';
}
```
